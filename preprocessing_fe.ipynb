```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading dataset
df = pd.read_csv("data/raw/housing.csv")

df.shape
```

```python
# Check missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
missing_values_percentage = df.isnull().sum() / len(df) * 100
missing_values_percentage[missing_values_percentage > 0].sort_values(ascending=False)
```

Remove column with more than 40% missing value

```python
threshold = 40

cols_to_drop = missing_values_percentage[missing_values_percentage > threshold].index

df = df.drop(columns=cols_to_drop)

df.columns
```

```python
# Fill missing numerical values with median
num_cols = df.select_dtypes(include=['number']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing categorical values with mode
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

# Verify missing values are handled
df.isnull().sum().sum()
```

# Feature Engineering

```python
# Total SF (Total Square Footage)
df['Total SF'] = df['Total Bsmt SF'] + df['1st Flr SF'] + df['2nd Flr SF']

# House Age, Remodel Age & Garage Age
df['House Age'] = df['Yr Sold'] - df['Year Built']
df['Years Since Remodel'] = df['Yr Sold'] - df['Year Remod/Add']
df['Garage Age'] = df['Yr Sold'] - df['Garage Yr Blt']

# Total Bathrooms
df['Total Bathrooms'] = df['Full Bath'] + (df['Half Bath'] * 0.5) + df['Bsmt Full Bath'] + (df['Bsmt Half Bath'] * 0.5)

# Total Porch Area
df['Total Porch SF'] = df['Wood Deck SF'] + df['Open Porch SF'] + df['Enclosed Porch'] + df['3Ssn Porch'] + df['Screen Porch']
```

```python
df.shape
```

```python
# 5. Encoding Ordinal Variables
qual_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1}
ordinal_cols = ['Exter Qual', 'Kitchen Qual', 'Bsmt Qual', 'Bsmt Cond']
df[ordinal_cols] = df[ordinal_cols].apply(lambda col: col.map(qual_map).fillna(0))
```

```python
df.shape
```

```python
# 6. One-Hot Encoding for Categorical Variables
categorical_features = ['Neighborhood', 'MS Zoning', 'Bldg Type', 'House Style', 'Roof Style', 'Foundation', 'Sale Type']
df = pd.get_dummies(df, columns=categorical_features, drop_first=True)
```

```python
df.shape
```

```python
df.head()
```

```python
# Drop original columns used for feature engineering
drop_feature_cols = ['Total Bsmt SF', '1st Flr SF', '2nd Flr SF', 'Year Built', 'Year Remod/Add', 'Garage Yr Blt', 'Full Bath', 'Half Bath', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch']
df.drop(columns=drop_feature_cols, inplace=True)
```

```python
# Drop Order, PID, Mo Sold and Yr Sold as they are not useful for modeling
df = df.drop(columns=["Order", "PID", "Mo Sold", "Yr Sold"])
```

```python
df.shape
```

```python
from scipy.stats import skew

num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Compute skewness for each column
skewness = df[num_cols].apply(lambda x: skew(x.dropna()))

# Filter highly skewed features (skewness > 1)
skewed_features = skewness[skewness > 1].index
print("Highly skewed features:", skewed_features)
```

```python
import numpy as np

# Apply log transformation only on positively skewed features
for feature in skewed_features:
    df[feature] = np.log1p(df[feature])  # log(1 + x) to avoid log(0) errors
```

```python
# Compute the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Matrix of Ames Housing Features")
plt.show()
```

```python
#Identify & Remove Highly Correlated Features
corr_threshold = 0.85
corr_matrix = df.corr(numeric_only=True)
high_corr_pairs = []
high_corr_features = set()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > corr_threshold:
            colname1 = corr_matrix.columns[i]
            colname2 = corr_matrix.columns[j]
            high_corr_pairs.append((colname1, colname2, corr_matrix.iloc[i, j]))
            high_corr_features.add(colname2)

# Print correlated pairs
print("Highly Correlated Feature Pairs:")
for pair in high_corr_pairs:
    print(f"{pair[0]} <--> {pair[1]} (Correlation: {pair[2]:.2f})")

# Drop highly correlated features
df.drop(columns=high_corr_features, inplace=True)
```
